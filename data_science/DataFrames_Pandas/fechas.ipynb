## Fechas  (Parte I)

```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Nombre': ['Alicia','Pedro','Carlos'],
    'Fecha_nacimiento': ['10/25/2005','10/29/2002','01/01/2001']
})
```

```python
df
```

```python
df['Fecha_nacimiento'] = pd.to_datetime(df['Fecha_nacimiento'])
```

```python
df
```

```python
df['anyo'] = df['Fecha_nacimiento'].map(lambda x: x.strftime('%Y'))
df
```

```python
df['mes'] = df['Fecha_nacimiento'].map(lambda x: x.strftime('%m'))
df
```

```python
df['dia'] = df['Fecha_nacimiento'].map(lambda x: x.strftime('%d'))
df
```
