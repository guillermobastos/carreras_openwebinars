## Gorup by

```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Company": ["GOOG", "GOOG", "MSFT", "MSFT", "FB", "FB"],
    "Person": ["Sam", "Charlie", "Sam", "Vanessa", "Carl", "Charlie"],
    "Sales": [200, 120, 340, 124, 243, 350],
    "SalesLastYear": [150, 170, 500, 100, 240, 300],
}
data
```

```python
data = pd.DataFrame(data)
data
```

```python
# Queremos calcular las ventas por company
porempresa = data.groupby('Company')
porempresa
```

```python
# Media de las ventas por compañía 
porempresa.mean([['Sales', 'SalesLastYear']])
```

```python
# Conteos, máximos y mínimos
data.groupby('Person').count()
```

```python
data.groupby('Person').max()
```

```python
data.groupby('Person').min()
```

```python
# Funcion describe
data.groupby('Company').describe()
```

```python
data.groupby('Company').describe().transpose()
```

```python
data.groupby('Company').describe().transpose()['FB']
```

```python
# Dos variables de agrupación
data.groupby(['Company','Person']).sum()
```
