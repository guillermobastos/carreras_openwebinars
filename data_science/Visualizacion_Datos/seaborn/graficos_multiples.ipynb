## Gráficos múltiples

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('titanic')
data.head(3)
```

```python
g = sns.FacetGrid(data = data, col = 'sex')
g.map(sns.histplot, 'age')
```

```python
g = sns.FacetGrid(data = data, col = 'pclass', hue = 'sex')
g.map(sns.scatterplot, 'age', 'fare')
g.add_legend()
```

```python
g = sns.FacetGrid(data = data, col = 'pclass')
g.map(sns.scatterplot, 'age', 'fare', color = 'green', s = 60)

for ax in g.axes.flat:
    ax.axhline(y = 300)
    
g.set(xlim=(0, 80), ylim=(0, 400))
```

```python
g = sns.PairGrid(data = data, vars = ['pclass', 'age', 'fare'], hue = 'sex')
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
```

```python
g = sns.PairGrid(data = data, vars = ['pclass', 'age', 'fare'], hue = 'sex')
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```

```python
sns.pairplot(data = data, vars = ['pclass', 'age', 'fare'], hue = 'sex', kind = 'scatter')
```
