```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('titanic')
data.head(2)
```

### Catplot
Representaci칩n predeterminada de los datos en s eaborn es catplot().
Usa un diagrama de dispersi칩n o puntos

```python
sns.catplot(x = 'pclass', y = 'age', data = data, height=7, palette='husl')
```

```python
sns.catplot(x = 'pclass', y = 'age', data = data, height=7, kind = 'swarm', palette='husl', hue='sex')
```

### Boxplots

```python
fig, ax = plt.subplots(figsize = (15, 10))
sns.boxplot(x = 'pclass', y = 'age', data = data, ax = ax, palette='husl')
```

```python
sns.catplot(x = 'pclass', y = 'age', data = data, height=7, kind = 'box', palette='husl', hue='sex', orient='v')
```

```python
sns.catplot(x = 'pclass', y = 'survived', data = data, height=7, kind = 'bar', palette='husl')
```

```python
sns.catplot(x = 'pclass', y = 'survived', data = data, height=7, kind = 'bar', palette='husl', estimator=np.count_nonzero)
```
