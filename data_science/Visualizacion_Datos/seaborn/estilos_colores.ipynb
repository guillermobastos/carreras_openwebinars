## Estilos y colores

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('titanic')
data.head(8)
```

```python
sns.boxplot(x = 'pclass', y = 'age', data = data)
```

```python
sns.set_style('whitegrid')
sns.boxplot(x = 'pclass', y = 'age', data = data)
```

```python
sns.boxplot(x = 'pclass', y = 'age', data = data)
sns.despine(left = True, offset=20, trim=True)
```

```python
sns.set_theme()
```

```python
sns.set_context()
```

```python
sns.set_context('talk')
sns.boxplot(x = 'pclass', y = 'age', data = data)
```

```python
sns.set_context('notebook', font_scale=0.9)
sns.boxplot(x = 'pclass', y = 'age', data = data, linewidth=3.5)
```

```python
sns.set_context('talk')
```

```python
sns.boxplot(x = 'pclass', y = 'age', data = data, palette=sns.color_palette("husl", 8))
```

```python
# Paired
sns.boxplot(x = 'pclass', y = 'age', data = data, palette=sns.color_palette("Set2", 8))
```

```python
numeric_data = data.select_dtypes(include='number')
sns.heatmap(numeric_data.corr(), cmap='Blues')
```
