## Gráficos de regresión

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
data = sns.load_dataset('tips')
data.head(10)
```

```python
fig,ax = plt.subplots(figsize = (10, 5))
sns.regplot(x = 'tip', y = 'total_bill', data = data, ax=ax, ci = None)
```

```python
fig,ax = plt.subplots(figsize = (10, 5))
sns.regplot(x = 'tip', y = 'total_bill', data = data, ax=ax, ci = True, line_kws={'color':'red'}, color = 'green')
```

```python
sns.lmplot(x = 'tip', y= 'total_bill', data = data, height = 6)
```

```python
sns.lmplot(x = 'tip', y= 'total_bill', data = data, height = 6, hue = 'smoker')
```

```python
sns.lmplot(x = 'size', y= 'total_bill', data = data, height = 6, order = 2)
```

```python
sns.lmplot(x = 'tip', y= 'total_bill', hue = 'smoker', data = data, height = 7, order = 2, col = 'sex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=data, kind = 'reg')
```
