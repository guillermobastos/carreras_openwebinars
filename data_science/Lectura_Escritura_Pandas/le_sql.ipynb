# SQL

* `to_sql`:
* `if_exists`: `{‘fail’, ‘replace’, ‘append’}`, defecto `‘fail’`

```python
import pandas as pd
import sqlite3
```

```python
data_folder_in = "data/in/{}"
data_folder_out = "data/out/{}"

database = 'chinook.db'
```

```python
cnx = sqlite3.connect(data_folder_in.format(database))
```

```python
df = pd.read_sql_query("SELECT * FROM playlists", cnx)
df.head(10)
```

```python
df['Name'] = df['Name'].str.lower()
df.head(10)
```

```python
df.to_sql('playlists', con=cnx, if_exists='append')
```

```python
df.to_sql('playlists', con=cnx, if_exists='append', index=False)
```

```python
df['PlaylistId'] = df['PlaylistId'] + 20
df
```

```python
df.to_sql('playlists', con=cnx, if_exists='append', index=False, index_label='PlaylistId')
```

```python
df = pd.read_sql_query("SELECT * FROM playlists", cnx)
df
```

```python
df['Name'] = df['Name'].str.lower()
df.head(10)
```

```python
df.to_sql('playlists', con=cnx, if_exists='replace', index=False, index_label='PlaylistId')
```

```python
df = pd.read_sql_query("SELECT * FROM playlists", cnx)
df
```

```python
df = pd.read_sql_query(
    """
    SELECT * FROM playlists p, playlist_track pt
    WHERE p.PlaylistId = pt.PlaylistId
""",
    cnx,
)
df
```

```python
df.to_csv(data_folder_out.format('chinook.csv'), index=False, sep=";", decimal=",")
```
