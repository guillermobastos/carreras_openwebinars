## qcut y cut

```python
pd.qcut(data['price'], q=5)
```

```python
pd.qcut(data['price'], q=5).value_counts()
```

```python
pd.cut(data['bedrooms'], bins=2)
```

```python
pd.cut(data['bedrooms'], bins=2).value_counts()
```

```python
pd.cut(data['bedrooms'], bins=[-1,3,8,10], labels=['menos de 3', 'entre 3 y 8', 'entre 8 y 10'])
```

```python
data.sort_values(by='bedrooms', ascending=True, inplace=False)
```
