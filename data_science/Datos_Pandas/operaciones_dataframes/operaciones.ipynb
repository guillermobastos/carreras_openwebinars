## Operaciones entre DataFrames

```python
import pandas as pd
import numpy as np

from pathlib import Path

data_path = Path('../estructura_datos/data')
data = pd.read_csv(data_path/'house.csv')
data.head(2)
```

```python
df1 = data.iloc[0:10]
df1.shape
```

```python
df2 = data.iloc[-10:] # coge las 10 Ãºltimas filas
df2.shape
```

```python
df = pd.concat([df1, df2], axis=0)
df
```

```python
df = pd.concat([df1, df2], axis=0, keys=['head','tail'])
df
```

```python
df.loc['tail']
```

```python
df1 = data.loc[0:10,['price','bedrooms']]
df1
```

```python
df2 = data.loc[0:10,['city','statezip','country']]
df2
```

```python
df = pd.concat([df1, df2], axis=1)
df
```
