## Visualización

```python
import pandas as pd
import numpy as np

import warnings

from pathlib import Path

import matplotlib.pyplot as plt

data_path = Path('../estructura_datos/data')
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv(data_path/'house.csv')
data.head(2)
```

```python
df = data.drop_duplicates(subset = ['date'], keep='first')
df.shape
```

```python
df.index = df['date']
df = df.drop(labels = ['date'], axis = 1)
df.head()
```

### Gráfico de lineas

```python
df[['price']].plot(figsize = (20, 8))
```

```python
df[['bathrooms','floors']].plot(kind='bar', figsize=(20,8))
```

```python
df[['bathrooms']].plot(kind = 'bar', figsize = (20, 8))
```

```python
df[['bathrooms', 'floors']].plot(kind = 'bar', figsize = (20, 8))
```

### Gráfico de puntos

```python
df.plot(kind = 'scatter', x = 'yr_built', y = 'sqft_above', figsize = (10, 8))
```

### Gráfico de cajas

```python
df[['bathrooms', 'floors']].plot(kind = 'box', figsize = (10, 8))
```

```python
df['price'].plot(kind = 'box', figsize = (10, 8))
```

### Gráfico de tarta

```python
df['condition'].value_counts(normalize = True).plot(kind = 'pie', figsize = (10, 8), legend = True)
```
