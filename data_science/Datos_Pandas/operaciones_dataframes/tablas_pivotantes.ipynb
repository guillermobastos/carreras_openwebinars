## Tablas pivotantes

```python
import pandas as pd
import numpy as np

from pathlib import Path

data_path = Path('../estructura_datos/data')
df = pd.read_csv(data_path/'pivot_data.csv')
df.head(10)
```

```python
df[df['variable'] == "B"]
```

```python
_ = df.pivot(index='date', columns='variable', values='value')
_
```

```python
_ = df.pivot(index='variable2', columns='variable', values='date')
_
```

```python
_ = df.pivot(index=['date','variable'], columns='variable2', values='value')
_
```

```python
_.index
```

```python
_ = df.pivot(index=['date', 'variable'], columns='variable2', values='value')
_
```

### Stack y unstack

```python
stack = _.stack().to_frame()
stack
```

```python
stack.unstack(level = 0)
```

```python
stack.unstack(level = 1)
```

```python
stack.unstack(level=['date','variable'])
```

### melt

```python
data_cal = pd.DataFrame(
    {
        "nombre": ["Antonio", "Maria"],
        "apellido": ["Lopez", "Jimenez"],
        "nota": [7, 8.9],
        "num_asignaturas": [3, 5],
    }
)
data_cal
```

```python
data_cal.melt(id_vars=['nombre','apellido'])
```
