### Agregaciones

```python
data.groupby('condition').agg({
    'price' : 'mean',
    'bathrooms' : 'median',
    'waterfront' : 'max',
    'floors' : 'min'
})
```

```python
data.groupby('condition').agg({
    'price' : 'mean',
    'bathrooms' : 'median',
    'waterfront' : 'max',
    'floors' : 'min'
})
```

```python
agrupacion = data.groupby('condition').agg({
    'price' : ['mean', 'min', 'max'],
    'bathrooms' : 'median',
    'waterfront' : 'max',
    'floors' : ['min', 'max']
})
agrupacion
```

```python
agrupacion['price']['max']
```

```python
agrupacion.stack(level=0, future_stack=True)
```

```python
def media_mayores_1millon(x):
    return np.mean(x[x > 1_000_000])
```

```python

data[['price','condition']].groupby('condition')['price'].apply(media_mayores_1millon)
```
