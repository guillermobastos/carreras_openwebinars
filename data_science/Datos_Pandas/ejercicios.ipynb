### Ejercicio 4

```python
# Convierte la columna 'date' en un tipo fecha
data['date'] = pd.to_datetime(data['date'], yearfirst=True)
data['date']
```

```python
# Genera 3 nuevas columnas fruto de extraer el año, mes y día 
data['year'] = data['date'].dt.year
data['month'] = data['date'].dt.month
data['day'] = data['date'].dt.day
```

```python
# Cual es el precio medio por mes?
data.set_index(data['date']).resample('ME')['price'].mean()
```

```python
# Crea una nueva columna fruto de sumar 20 dias a la columna fecha
td = pd.Timedelta('20 days')
td
```

```python
data['date2'] = data['date'] + td
data[['date', 'date2']].head()
```
