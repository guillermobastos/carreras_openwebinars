```python
import pandas as pd
```

```python
df_raw = pd.read_csv('data/bikes2018.csv',
                     sep=',', 
                     encoding='latin2', 
                     parse_dates=['Date'], 
                     dayfirst=True, 
                     index_col='Date')
```

```python
df_raw.head(3)
```

```python
df_raw['Berri1']
```

```python
df_raw.describe()
```

```python
df_selected = df_raw[['Berri1','Boyer','Boyer 2','Parc','PierDup','Viger']]
```

```python
df_selected.head(5)
```

```python
df_selected.fillna(value = 0)
```

```python
df_selected.sum()
```

```python
df_selected.sum().max()
```

```python
df_selected.sum().argmax()
```

```python
df_selected.sum().plot(kind='bar')
```

```python
df_selected.reset_index(inplace=True)
```

```python
df_selected.head(3)
```

```python
df = df_selected.copy()
```

```python
df.head()
```

```python
df['Date']
```

```python
df['Year'] = df['Date'].dt.year
```

```python
df['Month'] = df['Date'].dt.month
```

```python
df.head()
```

```python
df.fillna(0, inplace=True)
df.head(100)
```

```python
df.dtypes
```

```python
df.drop(columns=['Date']).groupby('Month').sum()
```

```python
df['Total'] = df['Berri1'] + df['Boyer'] + df['Boyer 2'] + df['Parc'] + df['PierDup']
```

```python
df.head()
```

```python
df.drop(columns=['Date']).groupby('Month').sum()
```

```python
df[['Month','Total']].groupby('Month').sum()
```

```python
df[['Month','Total']].groupby('Month').sum().plot(kind='bar')
```
